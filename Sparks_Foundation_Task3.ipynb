 ## Data Visualization

```python
# Graph between Category and Sub-Category

plt.figure(figsize = (20, 15))
plt.bar('Sub-Category', 'Category', data = df)
plt.show()
```

```python
print(df['State'].value_counts()) #Count of different States
```

```python
#Graph of number of State

plt.figure(figsize = (20, 10))
sns.countplot(x = df['State'])
plt.xticks(rotation = 90)
plt.show()
```

```python
print(df['Sub-Category'].value_counts())
```

```python
#Graph of number of Sub-Category

sns.countplot(x = df['Sub-Category'])
plt.xticks(rotation = 90)
plt.show()
```

```python
Profit_plot = (ggplot(df, aes(x='Sub-Category', y='Profit', fill='Sub-Category')) + geom_col() + coord_flip()
+ scale_fill_brewer(type='div', palette="Spectral") + theme_classic() + ggtitle('Pie Chart'))

display(Profit_plot)
```

```python
#Graph of number of Sub-Category

sns.countplot(x = df['Sub-Category'])
plt.xticks(rotation = 90)
plt.show()
```

```python
ggplot(df, aes(x='Ship Mode', fill = 'Category')) + geom_bar(stat = 'count')
```

```python
#Graph between Discount and Profit

sns.lineplot(x = df['Discount'], y = df['Profit'], data = df, color = 'red', label = 'Discount')
plt.legend()
plt.grid()
```

```python
figsize=(15,10)
sns.pairplot(df,hue='Sub-Category')
plt.show
```

```python
#Sales grouping based on State

Sales = df.groupby("State")["Sales"].sum()
Sales.plot.bar(figsize = (15, 10))
plt.show()
```

```python
#Category grouping based on sales and profit

Category = df.groupby("Category")["Profit","Sales"].sum()
Category.plot.bar(figsize = (15, 10))
plt.show()
```
